## 1. Importing Necessary Libraries ##

```python
#Importing necessary libraries
import pandas as pd
# import numpy as np
# from os import path
# from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt

%matplotlib inline
```

## 2. Importing Data and Pre-Processing ##

```python
#Importing data from respective csv files into Data Frames
data_men = pd.read_csv('../data/AskMen.csv')
data_women = pd.read_csv('../data/AskWomen.csv')

#Arranging the text content of our Data Frames into string objects for further processing
text_men = " ".join(content for content in data_men['text'])
text_women = " ".join(content for content in data_women['text'])
```

## 3. Plotting ##

### 3.1 Plotting 'AskMen' WordCloud ###

```python
#Arranging standard STOPWORDS into a set for further processing
stopwords = set(STOPWORDS)

#Instantiating a world cloud
wordcloud_men = WordCloud(stopwords = stopwords,
                          scale = 20,
                          max_font_size=200, 
                          max_words=300, 
                          background_color="white",
                         relative_scaling=.5,
                         colormap = 'winter',
                         normalize_plurals=True,
                         min_word_length=3).generate(str(text_men))

#Plotting the wordcloud
plt.figure(figsize=[20,20])
plt.imshow(wordcloud_men, interpolation='bilinear')
plt.axis("off")
plt.show()

#Saving the wordcloud into a png file to be used in illustraions
wordcloud_men.to_file('../images/wordcloud_men.png');
```

### 3.2 Plotting 'AskWomen' WordCloud ###

```python
#Instantiating a world cloud
wordcloud_women = WordCloud(stopwords = stopwords,
                          max_font_size=200, 
                          max_words=300, 
                          background_color="white",
                         relative_scaling=.5,
                         colormap = 'autumn',
                         normalize_plurals=True,
                         min_word_length=3
                            ).generate(str(text_women))

#Plotting the wordcloud
plt.figure(figsize=[20,20])
plt.imshow(wordcloud_women, interpolation='bilinear')
plt.axis("off")
plt.show()

#Saving the wordcloud into a png file to be used in illustraions
wordcloud_women.to_file('../images/wordcloud_women.png');
```
